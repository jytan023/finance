# Compute drawdowns

```python
import pandas as pd
import numpy as np
```

```python
returns =pd.read_csv('data/Portfolios_Formed_on_ME_monthly_EW.csv', 
                     header=0, index_col = 0, parse_dates = True, na_values =-99.99)
columns = ['Lo 10', 'Hi 10']
returns = returns[columns]
returns = returns/100
returns.columns = ['Small Cap', 'Large Cap']
returns.index = pd.to_datetime(returns.index, format = '%Y%m').to_period('M')
returns.head()
```

```python
returns.plot.line()
```

# Compute Drawdowns
1. Compute wealth index
2. compute previous peaks
3. Compute drawdown - which is the wealth as a percentage of the previous peak

```python
# value of portfolio over time with $1000
wealth_index = (1+returns["Large Cap"]).cumprod()
```

```python
wealth_index.head()
wealth_index.plot.line()
```

```python
# previous peaks
prev_peaks = wealth_index.cummax()
prev_peaks.plot.line()
```

```python
drawdown=(wealth_index-prev_peaks)/prev_peaks
drawdown.plot()
```

```python
def drawdowns(return_series: pd.Series):
    """
    Takes a time series asset returns
    Computes and returns dataframe that contains:
    1. The wealth index
    2. The previous peaks
    3. Percentage drawdowns
    """
    wealth_index = (1+return_series).cumprod()
    prev_peaks = wealth_index.cummax()
    drawdown=(wealth_index-prev_peaks)/prev_peaks
    return(pd.DataFrame({
        'Wealth' : wealth_index,
        'Peaks' : prev_peaks,
        'Drawdown' : drawdown
    }))
```

```python
table = drawdowns(returns['Small Cap'])
```

```python
value = table['Peaks'].value_counts().idxmax()
duration = table['Peaks'].value_counts().max()
date = table[table['Wealth']==value].index
```
